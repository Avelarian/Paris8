```python
import sys
import numpy as np
np.set_printoptions(threshold=sys.maxsize)
import pandas as pd
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
mouse_data_frame = pd.read_csv('mouse.txt', sep=' ', header=None)

plt.scatter(mouse_data_frame[0], mouse_data_frame[1], color='black')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
from sklearn.ensemble import IsolationForest

model = IsolationForest(n_estimators=500, contamination=0.02)
model.fit(mouse_data_frame)

plt.scatter(mouse_data_frame[0], mouse_data_frame[1], c = model.predict(mouse_data_frame))
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
creditcard_data_frame = pd.read_csv('creditcard.csv', sep=',')
creditcard_data_frame = creditcard_data_frame.drop(columns=['Time'])

creditcard_data_frame_class_0 = creditcard_data_frame[creditcard_data_frame['Class'] == 0]
creditcard_data_frame_class_0 = creditcard_data_frame_class_0.sample(n=5000)
creditcard_data_frame_class_1 = creditcard_data_frame[creditcard_data_frame['Class'] == 1]

creditcard_data_frame_new = pd.concat([creditcard_data_frame_class_0, creditcard_data_frame_class_1])

X_credit = creditcard_data_frame_new.iloc[:, 0:28].values
labels_class_credit = creditcard_data_frame_new.iloc[:, 29].values
```

```python
IF=IsolationForest(n_estimators=2000)
IF.fit(X_credit)
decision=IF.decision_function(X_credit)

plt.figure(figsize=(8, 6))
plt.hist(decision, bins=50);
```

```python
deicisionfinale=np.copy(decision)
deicisionfinale[decision<0]=1
deicisionfinale[decision>=0]=0
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(labels_class_credit,deicisionfinale)
```
