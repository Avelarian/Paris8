```python
import sys
import numpy as np
np.set_printoptions(threshold=sys.maxsize)
import pandas as pd
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
mouse_data_frame = pd.read_csv('mouse.txt', sep=' ', header=None)

plt.scatter(mouse_data_frame[0], mouse_data_frame[1], color='black')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
from sklearn.ensemble import IsolationForest

model = IsolationForest(n_estimators=500, contamination=0.02)
model.fit(mouse_data_frame)

plt.scatter(mouse_data_frame[0], mouse_data_frame[1], c = model.predict(mouse_data_frame))
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
creditcard_data_frame = pd.read_csv('creditcard.csv', sep=',')
creditcard_data_frame = creditcard_data_frame.drop(['Time'])

creditcard_data_frame_class_0 = creditcard_data_frame[creditcard_data_frame['Class'] == 0]
creditcard_data_frame_class_0 = creditcard_data_frame_class_0.sample(n=5000)

creditcard_data_frame_class_1 = creditcard_data_frame[creditcard_data_frame['Class'] == 1]

credit_data_frame = pd.concat([creditcard_data_frame_class_0, creditcard_data_frame_class_1])
```

```python
model = IsolationForest(n_estimators=5492, contamination=0.09)
model.fit(creditcard_data_frame)

plt.scatter(creditcard_data_frame['Amount'], creditcard_data_frame['Class'], c=model.predict(creditcard_data_frame))
plt.show()
```
